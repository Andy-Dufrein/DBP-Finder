```python
train = pd.read_csv('../data/embeddings/input_csv/train_p2.csv')
```

```python
train.head()
```

```python
train_with_go = go_df.merge(train, on='identifier')
```

```python
neg_cl = train_with_go[train_with_go['label']==0]
```

```python
def check_go_term(entry):
    dna_bind = 'GO:0003677' in entry
    rna_bind = 'GO:0003723' in entry
    nucl_bind = 'GO:0003676' in entry
    return any([dna_bind, rna_bind, nucl_bind])
```

```python
len(neg_cl)
```

Смотрю негативный класс, кто имеет аннотацию днк, рнк, нк связывания

```python
neg_cl_with_na = neg_cl[neg_cl['Gene Ontology (molecular function)'].isna()]
```

```python
neg_cl_without_na = neg_cl[neg_cl['Gene Ontology (molecular function)'].notna()]
```

```python
protein_id = neg_cl_without_na[neg_cl_without_na['Gene Ontology IDs'].apply(check_go_term)]['identifier']
```

```python
train_upd = train.loc[~train['identifier'].isin(protein_id)]
```

```python
train_upd.to_csv('../data/processed/train_upd.csv', index=False)
```
